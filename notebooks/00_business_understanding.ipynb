## 1. Contexto del Problema

En mercados, plazas de abastos y agroindustrias de todo el mundo, la clasificación de productos frescos —tomates, manzanas, naranjas, bananos, limas y granadas, entre otros— se realiza de forma manual. Un operario recorre las bandas de empaque inspeccionando visualmente cada unidad y asignando, con base en su experiencia, una categoría de calidad y un rango de tamaño. Este proceso presenta cuatro limitaciones estructurales que afectan directamente la rentabilidad y la competitividad del sector:

1. **Subjetividad:** el criterio de clasificación varía entre operarios y a lo largo de la jornada laboral, lo que genera inconsistencia en los lotes y devoluciones de producto por parte de los compradores.
2. **Velocidad:** un inspector humano puede revisar entre 400 y 800 unidades por hora; una línea automatizada puede superar las 10.000 unidades en el mismo tiempo.
3. **Fatiga:** la inspección continua reduce la precisión a medida que avanza el turno, especialmente en condiciones de iluminación artificial.
4. **Pérdida económica:** el rechazo de lotes completos por mala clasificación y el desperdicio de frutas que no alcanzan el mercado representan, según la FAO, aproximadamente el 14% de los alimentos producidos a nivel global antes de llegar al consumidor.

La visión por computador ofrece una alternativa objetiva, reproducible y escalable. Un sistema que analice fotografías de frutas individuales y asigne automáticamente una categoría de calidad y una estimación de tamaño puede reducir el desperdicio, uniformizar los estándares de empaque y abaratar los costos de operación de las plantas de procesamiento.

## 2. Descripción Detallada del Problema

El problema central que aborda este proyecto es el desarrollo de un **sistema automático de clasificación de calidad y tamaño de frutas** basado en imágenes digitales, utilizando técnicas de aprendizaje automático (*machine learning*) y aprendizaje profundo (*deep learning*).

### 2.1 Entradas

El sistema recibe como entrada **imágenes estáticas** (fotografías tomadas con cámara web o celular) de una fruta o verdura individual sobre un fondo simple y uniforme. Estas imágenes pueden provenir de:

- El conjunto de datos público *Fruit Quality Classification* disponible en Kaggle.
- Imágenes adicionales recolectadas en plazas de mercado, supermercados y hogares.

### 2.2 Salidas

El sistema debe producir dos predicciones independientes para cada imagen:

| Tarea | Clases de salida | Naturaleza |
|---|---|---|
| Clasificación de calidad | `good` (buena), `regular` (intermedia), `bad` (dañada) | Clasificación multiclase |
| Clasificación de tamaño | `pequeño`, `mediano`, `grande` | Clasificación multiclase |

### 2.3 Complejidad del problema

Este es un **problema complejo de ingeniería** por las siguientes razones:

- Las clases de calidad no son linealmente separables en el espacio RGB ni en el espacio HSV cuando se usan únicamente promedios estadísticos de color.
- El conjunto de datos presenta un **desbalance de clases crítico**: el Índice de Imbalance (IR) supera el umbral de 3 en cuatro de las seis frutas incluidas, llegando a IR = 14.40 en granadas.
- Las imágenes provienen de fuentes heterogéneas con resoluciones muy distintas, lo que introduce ruido en las características geométricas absolutas.
- Los defectos de calidad (manchas, podredumbre, hematomas) son fenómenos locales de textura que no pueden ser capturados por descriptores estadísticos globales, requiriendo análisis espacial mediante redes convolucionales.

### 2.4 Formulación matemática

Sea $\mathbf{x}_i \in \mathbb{R}^9$ el vector de características extraídas de la $i$-ésima imagen, donde:

$$\mathbf{x}_i = [\text{area\_px},\ \text{aspect\_ratio},\ \text{coverage\_ratio},\ h_\mu,\ s_\mu,\ v_\mu,\ h_\sigma,\ s_\sigma,\ v_\sigma]$$

El problema de clasificación de calidad se formula como encontrar una función $f_{\text{calidad}}: \mathbb{R}^9 \to \{\text{good},\ \text{regular},\ \text{bad}\}$ y el de tamaño como $f_{\text{tamaño}}: \mathbb{R}^9 \to \{\text{pequeño},\ \text{mediano},\ \text{grande}\}$ que minimicen el error de clasificación sobre un conjunto de prueba independiente, medido mediante la métrica **F1-Score Macro**:

$$F1_{\text{macro}} = \frac{1}{|C|} \sum_{c \in C} \frac{2 \cdot P_c \cdot R_c}{P_c + R_c}$$

donde $C$ es el conjunto de clases, $P_c$ es la precisión de la clase $c$ y $R_c$ es el recall de la clase $c$.

## 3. Justificación de la Importancia del Problema

La importancia de automatizar la clasificación de frutas se puede argumentar desde cuatro dimensiones:

### 3.1 Dimensión económica
Según datos del Banco Mundial y la FAO, el desperdicio de alimentos en la etapa de poscosecha (clasificación, empaque y transporte) supone pérdidas económicas de más de USD 940 mil millones anuales a nivel global. En Colombia, el Ministerio de Agricultura reporta que entre el 30% y el 40% de la producción hortofrutícola se pierde antes de llegar al consumidor, principalmente por problemas de clasificación y empaque. Un sistema automático preciso puede reducir este desperdicio en forma directa.

### 3.2 Dimensión social
La automatización de la clasificación permite a los pequeños productores acceder a mercados de exportación con estándares estrictos de calidad (UE, EE.UU., Japón) que hoy les están vetados por la falta de capacidad de clasificación. Esto amplía sus ingresos y reduce la brecha entre productores de subsistencia y empresas agroindustriales.

### 3.3 Dimensión ambiental
Reducir el desperdicio de alimentos es una de las acciones con mayor impacto en la reducción de emisiones de CO₂. La producción de alimentos que terminan desechados genera gases de efecto invernadero asociados a todo el ciclo productivo (riego, fertilizantes, transporte) sin que exista beneficio alimentario.

### 3.4 Dimensión técnica
Este proyecto constituye un caso de aplicación real de los fundamentos de aprendizaje automático y visión por computador adquiridos durante el curso. La comparación entre modelos clásicos (SVM, XGBoost) y modelos de aprendizaje profundo (CNN) sobre el mismo conjunto de datos permite establecer criterios técnicos objetivos para la selección de arquitecturas en problemas reales de ingeniería.

## 4. Objetivos

### 4.1 Objetivo General

Desarrollar un sistema automático de clasificación de calidad y tamaño de frutas, basado en visión por computador y técnicas de aprendizaje automático y aprendizaje profundo, utilizando la metodología CRISP-DM, capaz de procesar imágenes digitales y producir predicciones precisas comparables con los estándares de la industria agroalimentaria.

### 4.2 Objetivos Específicos

1. **Comprender el dominio** (Business Understanding): Definir formalmente el problema, sus objetivos de negocio, criterios de éxito y restricciones, así como identificar stakeholders, riesgos e impactos potenciales.

2. **Analizar los datos** (Data Understanding): Realizar un análisis exploratorio exhaustivo del conjunto de datos que incluya distribución de clases, detección de desbalance, análisis geométrico, cromático y espacial de las imágenes, identificando los patrones relevantes para el modelado.

3. **Preparar los datos** (Data Preparation): Implementar un pipeline de preprocesamiento que comprenda detección automática de frutas (YOLOv8 + métodos clásicos de segmentación), extracción de características geométricas y cromáticas en espacio HSV, generación automática de etiquetas de tamaño mediante percentiles, y balanceo de clases mediante aumento de datos.

4. **Entrenar y ajustar modelos** (Modeling): Implementar y optimizar al menos dos modelos clásicos de aprendizaje automático (SVM y XGBoost) y una red neuronal convolucional (CNN), comparando su desempeño sobre tareas de clasificación de calidad y tamaño.

5. **Evaluar los modelos** (Evaluation): Medir el desempeño de cada modelo usando métricas robustas frente al desbalance (F1-Score Macro, MCC, Coeficiente Kappa de Cohen, AUC-ROC), comparar los resultados con una línea base (*Dummy Classifier*) y determinar el modelo óptimo para cada tarea.

6. **Desplegar el sistema** (Deployment): Integrar los modelos entrenados en una interfaz gráfica funcional que permita al usuario cargar una imagen o capturarla en tiempo real desde la cámara y recibir las predicciones de calidad y tamaño.

## 5. Criterios de Éxito

Un proyecto de clasificación automática puede definir el éxito desde la perspectiva del negocio y desde la perspectiva técnica.

### 5.1 Criterios técnicos

| Criterio | Umbral mínimo | Umbral objetivo |
|---|---|---|
| F1-Score Macro en clasificación de calidad | ≥ 0.85 | ≥ 0.93 |
| F1-Score Macro en clasificación de tamaño | ≥ 0.90 | ≥ 0.95 |
| AUC-ROC Macro | ≥ 0.90 | ≥ 0.97 |
| MCC (Coeficiente Matthews) | ≥ 0.75 | ≥ 0.90 |
| Mejora sobre baseline (*Dummy Classifier*) | ≥ 50 puntos porcentuales en F1-Macro | — |
| Tiempo de inferencia por imagen | < 5 segundos | < 1 segundo |

### 5.2 Criterios del negocio

| Criterio | Descripción |
|---|---|
| Tasa de falsos negativos en clase `bad` | Debe ser baja (< 10%) para evitar que frutas dañadas pasen a empaque |
| Tasa de falsos positivos en clase `good` | Debe ser controlada para no rechazar fruta apta injustamente |
| Interfaz funcional | El sistema debe permitir cargar una imagen y obtener resultado en < 5s |
| Documentación del código | El código debe estar comentado y reproducible |

### 5.3 Criterio de selección del modelo final

Se construirá una **matriz de decisión** ponderando cuatro métricas (F1-Macro: 35%, MCC: 25%, Kappa: 20%, Balanced Accuracy: 20%) para cada tarea. El modelo con mayor puntuación ponderada será seleccionado como el mejor para cada tarea.

## 6. Restricciones y Supuestos

### 6.1 Restricciones

| Tipo | Restricción | Impacto |
|---|---|---|
| **Datos** | El conjunto de datos de referencia está limitado a 6 tipos de fruta (manzana, banano, lima, naranja, granada, guayaba) | Los modelos no generalizarán a frutas no incluidas sin reentrenamiento |
| **Imagen** | Las imágenes de entrada deben capturarse sobre fondo simple y uniforme | Fondos complejos reducirán la precisión de detección |
| **Resolución** | El dataset mezcla imágenes de muy distintas resoluciones nativas | Obliga a normalizar el área mediante porcentaje de cobertura, no píxeles absolutos |
| **Recursos computacionales** | El entrenamiento de la CNN se realiza en entorno académico sin GPU dedicada de alta gama | Limita la profundidad de la arquitectura CNN y el número de épocas |
| **Tiempo** | El proyecto se desarrolla en el marco de un semestre académico | Limita la exploración exhaustiva de arquitecturas y la recolección de datos propios |
| **Formato de entrada** | El sistema procesa imágenes estáticas, no video en tiempo real (excepto extensión opcional) | No aplica directamente a líneas de empaque de alta velocidad sin adaptación |

### 6.2 Supuestos

1. Se asume que las etiquetas de calidad (`good`, `regular`, `bad`) en el dataset de Kaggle son consistentes y representan el juicio de expertos en el dominio.
2. Se asume que las imágenes recolectadas por los estudiantes son representativas de la variabilidad real de cada fruta en el mercado local colombiano.
3. Se asume que el espacio de color HSV es una representación más robusta al ruido de iluminación que el espacio RGB para la tarea de clasificación de calidad.
4. Se asume que el área normalizada (porcentaje de cobertura de imagen) es un indicador válido del tamaño relativo de la fruta, bajo la condición de una distancia de captura aproximadamente constante.
5. Se asume que la partición 80/20 (entrenamiento/prueba) con estratificación preserva la distribución de clases y garantiza una evaluación representativa.

### 8.1 Análisis del riesgo R1: Desbalance de clases

El Índice de Imbalance se define como:

$$IR = \frac{\text{Clase Mayoritaria}}{\text{Clase Minoritaria}}$$

Un valor de $IR > 3$ se considera crítico porque indica que la clase minoritaria tiene menos de un tercio de las muestras de la mayoritaria. En nuestro dataset encontramos los siguientes valores antes del balanceo:

| Fruta | IR | Clase minoritaria | Observación |
|---|---|---|---|
| Granada (*pomegranate*) | 14.40 | regular | Crítico |
| Naranja | 5.59 | regular | Crítico |
| Manzana | ~3.2 | regular | Crítico |
| Limón (*lime*) | ~3.1 | regular | Crítico |
| Banano | 1.57 | — | Aceptable |
| Guayaba | ~1.8 | — | Aceptable |

## 9. Impacto Esperado de la Solución

### 9.1 Impacto técnico

El proyecto demuestra que es posible construir un clasificador automático de frutas con un desempeño superior al 93% de F1-Score Macro utilizando características visuales simples (área, proporción, estadísticas de color HSV) para los modelos clásicos, y directamente desde imágenes crudas para la CNN. Esto valida la viabilidad técnica de reemplazar la inspección visual humana en entornos industriales controlados.

### 9.2 Impacto económico

- **Reducción de desperdicio:** un clasificador con F1 > 0.93 en calidad puede reducir el rechazo de lotes completos por mala clasificación, disminuyendo las pérdidas poscosecha.
- **Reducción de costos laborales:** la automatización de la inspección libera personal para tareas de mayor valor agregado.
- **Acceso a mercados de exportación:** los productores colombianos podrían certificar la calidad de sus lotes con mayor facilidad.

### 9.3 Impacto social

- Mejora la equidad de acceso a tecnología para pequeños productores.
- Puede contribuir a la seguridad alimentaria al reducir el desperdicio en la cadena de suministro.
- Genera conocimiento transferible sobre metodología de aprendizaje automático aplicado al sector agropecuario.

### 9.4 Impacto ambiental

- **Positivo:** reducción del desperdicio alimentario y de las emisiones de CO₂ asociadas a alimentos que no llegan al consumidor.
- **Negativo (mitigable):** el entrenamiento de modelos de deep learning consume energía eléctrica significativa. Este impacto se mitiga usando EarlyStopping y entrenando modelos pequeños y eficientes.

### 9.5 Impacto global

La solución es generalizable a cualquier cultivo frutal con datos suficientes. Los modelos y pipelines desarrollados pueden ser reutilizados y adaptados para otros contextos culturales y regulatorios, y pueden escalar desde el nivel de pequeño productor hasta el de multinacional agroindustrial.
